```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use('classic')
imagen=Image.open("/home/esvelpe/Jupyter/imagenes/imagen_ejemplo_frida_bredesen.jpg")

plt.imshow(imagen)
```

```python
imagen_gr=imagen.convert('LA')
print(imagen_gr)
```

```python
imagen_mat=np.array(list(imagen_gr.getdata(band=0)),float)
print(imagen_mat)
```

```python
imagen_mat.shape = (imagen_gr.size[1],imagen_gr.size[0])
print(imagen_mat)
```

```python
print(imagen_mat.shape)
```

```python
plt.imshow(imagen_mat, cmap='gray')
plt.show()
```
